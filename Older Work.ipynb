```python
# plot the pairplot on the dataframe, with the outcome of diabetes as the color
# to differentiate between the classes in graphs
sns.pairplot(df_diabetes, hue='Outcome')
```

From the above pair plots we can conclude the following:

* When using hue variable in pairplot, we get density plot on the diagonals.
* People of younger age has lesser chance of diabetes than older age.
* Having a higher BMI indicates higher chances of diabetes.
* Having lower value of Glucose indicates lesser chance of diabetes.
* Having lower value of Pregnancy indicates lesser chance of diabetes.
* Combination of two features mostly overlap and therefore, more features are needed to classify if a person has a diabetes or not

```python
# calculate the correlation between different columns, excluding the Outcome column
df_diabetes_correlation = df_diabetes.drop('Outcome', axis=1).corr()
df_diabetes_correlation
```

```python
# plot the heatmap of the correlation dataframe
sns.heatmap(df_diabetes_correlation)
```

```python
# split the dataframe into row-column pairs and values, sort in descending order 
# and remove duplicates as same values exist in upper and lower triangles
correlation_values = df_diabetes_correlation.stack().sort_values(ascending=False).drop_duplicates().reset_index()

# remove those rows too which are self correlated, i.e. (Age, Age) = 1.0
correlation_values = correlation_values[~(correlation_values['level_0'] == correlation_values['level_1'])]
# the highest correlation is the first row
highest_correlation = correlation_values.iloc[0].values

# the lowest correlation is the last row
lowest_correlation = correlation_values.iloc[-1].values

# print the highest and lowest correlation and their respective columns
print('Highest correlation: ', highest_correlation)
print('Lowest correlation: ', lowest_correlation)
```

From above results, we can conclude: 
* Highest correlation is between Pregnancies and Age, as with increase in age, pregnancy count also increases and so does chances of diabetes.  
* Lowest correlation is between SkinThickness and Age, as it doesn't have any impact on each other or on diabetes

```python
# using crosstab, check a relation between the pregnancies and the diabetes outcome
# see if having pregranancy is correlated to diabetes or not
df_pregnancies_vs_diabetes = pd.crosstab(df_diabetes['Pregnancies'], df_diabetes['Outcome'])
df_pregnancies_vs_diabetes
```

From above, we can assume with lesser pregnancies, there is less chances of having diabetes, but with increase in pregnancies, chances of diabetes increases.
